## Basic Anomality Detection with DBSCAN (unsupervised)

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_moons
```

```python
# Create random dataset
X, y = make_moons(n_samples=750, noise=0.15)

# Plot generated data
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
# Initialise the DBSCAN model
dbscan = DBSCAN(eps=0.10)

# Train the model and make predictions on the input data
dbscan.fit_predict(X)
```

In the predicted labels, -1 represents outliers

```python
# Plot predicted data
plt.scatter(X[:,0], X[:,1], c=dbscan.labels_)
```
